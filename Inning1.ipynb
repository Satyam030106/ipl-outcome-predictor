```python
# extract innings==2 rows (unique per match) with their runs_target
df_innings2 = df[df['innings'] == 2][['match_id', 'batting_team', 'runs_target']].drop_duplicates()
# compute final score of the team batting in innings==1 per match
inning1_scores = df[df['innings'] == 1].groupby('match_id').agg({'team_runs': 'max'}).reset_index().rename(columns={'team_runs': 'inning1_score'})
df_new = pd.merge(df_innings2, inning1_scores, on='match_id')
```

```python
df_new
```

```python
df_new.drop(columns=['runs_target','batting_team'],inplace=True)
```

```python
df=df[df['innings']==1]
df.drop(columns=['ball_no'],inplace=True)
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df=df.merge(df_new,on='match_id')
```

```python
from sklearn.model_selection import train_test_split
X = final_df.drop(columns=['Final_score','match_id'])
y = final_df['Final_score']
X_check =final_df.drop(columns=['Final_score'])
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
X
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(X.iloc[:,:5])
category = ohe.categories_
trf = ColumnTransformer([
    ('trf',OneHotEncoder(categories=category,sparse_output=False,drop='first'),['batting_team','bowling_team','city','toss_winner','toss_decision'])
],remainder='passthrough')
```

```python
pipe = Pipeline([
    ('step1',trf),
    #('step2',RandomForestRegressor(n_jobs=-1))
    ("step2",XGBRegressor(n_estimators=2000, learning_rate=0.1, random_state=1))
])
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```
