```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("IPL.csv")
```

```python
df.head(10)
```

```python
df.drop(columns=['wicket_kind','player_out','fielders','date','match_type','event_name','bat_pos','batter','bowler','runs_batter','balls_faced'],inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(columns=['valid_ball','runs_extras','runs_not_boundary','extra_type','non_striker','non_striker_pos','review_batter',"team_reviewed",'review_decision',"umpire",'umpires_call','player_of_match','toss_winner',"toss_decision","venue","day",'month','year','season','gender',"team_type",'superover_winner','result_type','method','balls_per_over','overs','event_match_no',"stage","match_number",'new_batter','batter_runs','batter_balls','bowler_wicket','batting_partners','next_batter','striker_out','win_outcome'],inplace=True)
```

```python
df.head(5)
```

```python
df=df[df['innings']==2]
df.rename(columns={'match_won_by':'winner'},inplace=True)
df.drop(columns=['ball_no'],inplace=True)

```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.team_balls.unique()
```

```python

df.rename(columns={'team_runs':'current_score'},inplace=True)
```

```python
df.head(3)
```

```python
df['runs_left'] = df['runs_target']-df['current_score']+1
df["balls_left"] = 120 - df['team_balls']
```

```python

df.head(5)
```

```python
df['CRR']=df.current_score*6/(120-df.balls_left)
df['RRR']=df.runs_left*6/df.balls_left
df['wickets']=10- df.team_wicket
```

```python
df.head(5)
```

```python
df = df.reset_index()
```

```python
df.drop(columns='index',inplace=True)
```

```python
def is_win(f):
    winner = []
    for item,row in f.iterrows():
        if row.winner == row.batting_team:
            winner.append(1)
        else:
            winner.append(0)

    return winner
```

```python
df['winner']=is_win(df)
```

```python
df.head(5)
```

```python
final_df =df[['match_id','batting_team','bowling_team','city','runs_left','balls_left','wickets','runs_target','CRR','RRR','winner']]
```

```python
final_df.head(5)
```

```python
final_df.duplicated().sum()
```

```python
final_df =final_df.sample(final_df.shape[0])
```

```python
final_df.batting_team.unique()
```

```python
final_df['batting_team']=final_df['batting_team'].str.replace('Gujarat Lions','Gujarat Titans')
final_df['bowling_team']=final_df['bowling_team'].str.replace('Gujarat Lions','Gujarat Titans')

final_df['batting_team']=final_df['batting_team'].str.replace('Kings XI Punjab','Punjab Kings')
final_df['bowling_team']=final_df['bowling_team'].str.replace('Kings XI Punjab','Punjab Kings')

final_df['batting_team']=final_df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
final_df['bowling_team']=final_df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

final_df['batting_team']=final_df['batting_team'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
final_df['bowling_team']=final_df['bowling_team'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')

final_df['batting_team']=final_df['batting_team'].str.replace('Pune Warriors','Rising Pune Supergiant')
final_df['bowling_team']=final_df['bowling_team'].str.replace('Pune Warriors','Rising Pune Supergiant')

final_df['batting_team']=final_df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
final_df['bowling_team']=final_df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
final_df.batting_team.unique()
```

```python
teams_to_remove = ['Rising Pune Supergiant', 'Rising Pune Supergiants', 'Kochi Tuskers Kerala']
final_df = final_df[~final_df['batting_team'].isin(teams_to_remove) & ~final_df['bowling_team'].isin(teams_to_remove)]
```

```python
final_df.batting_team.unique()
```

```python
final_df.city.unique()
```

```python
final_df['city']=final_df['city'].str.replace('Bangalore','Bengaluru')
```

```python
final_df.info()
```

```python
final_df.dropna(inplace=True)
final_df =final_df[final_df.balls_left !=0]
```

```python
final_df.RRR.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
final_df.CRR.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.info()
```

```python
from sklearn.model_selection import train_test_split
X = final_df.drop(columns=['winner','match_id'])
y = final_df['winner']
X_check =final_df.drop(columns=['winner'])
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X
```

```python
y
```

```python
ohe = OneHotEncoder()
ohe.fit_transform(X.iloc[:,:3])
category = ohe.categories_
trf = ColumnTransformer([
    ('trf',OneHotEncoder(categories=category,sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='lbfgs',max_iter=1000))
    #('step2',RandomForestClassifier(n_jobs=-1))
   # ("step2",XGBClassifier(n_estimators=1000, learning_rate=0.1, random_state=1))
])
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```
